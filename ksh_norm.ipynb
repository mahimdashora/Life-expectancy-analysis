```python
import pandas as pd
import numpy as np
data=pd.read_csv('C:/Users/KSHITIZ/Desktop/Life_expectancy_cleaned.csv')
data
```

```python
new_data=data
new_data.drop(["Country","Status"],axis=1,inplace=True)
new_data.drop(data.columns[data.columns.str.contains('unnamed',case=False)],axis=1,inplace=True)
new_data
```

```python
print(new_data.mean(axis=0),"\n")
print(new_data.std(axis=0))
```

```python
from sklearn.preprocessing import StandardScaler
import statistics
scaler = StandardScaler() 
stand = scaler.fit_transform(new_data)
```

```python
std_data=pd.DataFrame(stand)
```

```python
print(std_data.mean(axis=0))
print(std_data.var())
```

```python
std_data=std_data.loc[:,19::1]
std_data.columns =['Life_Expectancy','Adult_Mortality','Infant_Deaths','Alcohol','Percentage','HepatitisB','Under_Five_Deaths','Polio','Tot_Exp','Diptheria','HIV/AIDS','GDP','Population','thinness_1to19_years','thinness_5to9_years','Income_Comp_Of_Resources','Schooling']
std_data #z score normalisation
```

```python
from sklearn import preprocessing 
minmax_scale = preprocessing.MinMaxScaler().fit(std_data)
minmax_df = minmax_scale.transform(std_data)
```

```python
minmax_df=pd.DataFrame(minmax_df)
```

```python

minmax_df.columns =['Life_Expectancy','Adult_Mortality','Infant_Deaths','Alcohol','Percentage','HepatitisB','Under_Five_Deaths','Polio','Tot_Exp','Diptheria','HIV/AIDS','GDP','Population','thinness_1to19_years','thinness_5to9_years','Income_Comp_Of_Resources','Schooling']
minmax_df #data set after normalisation
```

```python
print(minmax_df.mean(axis=0))
print(minmax_df.var())
```

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.gofplots import qqplot
fig,ax=plt.subplots(nrows=17,ncols=1,figsize=(10,50))
fig.subplots_adjust(hspace=1)
plt.suptitle("QQ Plots",x=0.5,y=0.92,size=30,weight="bold",color="darkblue")
all_col = ['Life_Expectancy','Adult_Mortality','Infant_Deaths','Alcohol','Percentage','HepatitisB','Under_Five_Deaths','Polio','Tot_Exp','Diptheria','HIV/AIDS','GDP','Population','thinness_1to19_years','thinness_5to9_years','Income_Comp_Of_Resources','Schooling']
for i in range(len(all_col)):
    qqplot(std_data[all_col[i]],line='s',ax=ax[i])
    ax[i].set_title(all_col[i])
 
plt.show()
```
